```python
import requests
import concurrent.futures
import time

# Define the API URL you want to test
api_url = "http://4fbe-35-229-192-149.ngrok-free.app" #replace this with your ngrok URL

# Define the number of requests to send concurrently for load balancing test
num_requests = 1000

# Function to send an API request
def send_request(url):
    try:
        response = requests.get(url)
        # You can analyze the response here if needed
        return response.status_code
    except Exception as e:
        return str(e)

# Function to perform load balancing test
def perform_load_balancing_test(api_url, num_requests):
    start_time = time.time()

    with concurrent.futures.ThreadPoolExecutor(max_workers=num_requests) as executor:
        # Start multiple requests concurrently
        future_to_url = {executor.submit(send_request, api_url): url for url in [api_url] * num_requests}

        for future in concurrent.futures.as_completed(future_to_url):
            url = future_to_url[future]
            status_code = future.result()
            print(f"Request to {url} returned status code: {status_code}")

    end_time = time.time()
    total_time = end_time - start_time
    print(f"Load balancing test took {total_time:.2f} seconds for {num_requests} requests")

if __name__ == "__main__":
    perform_load_balancing_test(api_url, num_requests)
```
